```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data_file = '../data/raw/hotel_bookings.csv'
df = pd.read_csv(data_file)
df.info()
```

```python
df.describe
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
# Check null values

null = pd.DataFrame({'Null Values' : df.isna().sum(), 'Percentage Null Values' : (df.isna().sum()) / (df.shape[0]) * (100)})
null
```

```python
# Handel null values

df['children'] = df['children'].fillna(df['children'].mean())

df['country'] = df['country'].fillna('Unknown Country')

df['agent'] = df['agent'].fillna(0)

df.drop(columns=['company'], inplace=True)
```

```python
# 

filter = (df.children == 0) & (df.adults == 0) & (df.babies == 0)
df[filter]
```

```python
# Check and handle records with negative ADR

print("ADR < 0 count:", (df['adr'] < 0).sum())
print("ADR = 0 count:", (df['adr'] == 0).sum())

df = df[df['adr'] >= 0]
```

```python
# Check mismatches in cancellation indicators

inconsistent = df[
    (df['is_canceled'] == 1) & (df['reservation_status'] != 'Canceled')
]
print("Inconsistent cancellations:", inconsistent.shape[0])

# If any found, drop them
df = df.drop(inconsistent.index)
```

```python
# Remove duplicates to recude bias

print("Duplicate rows:", df.duplicated().sum())

# Drop duplicates
df = df.drop_duplicates()
```

```python
df.to_csv("../data/processed/hotel_booking_clean.csv", index=False)
```
